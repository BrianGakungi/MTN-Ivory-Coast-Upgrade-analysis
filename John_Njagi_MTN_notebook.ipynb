MERGING df1 and df5 on cellid column


```python
df_merged = df1.merge(df5, how='outer', on=["CELL_ID"], left_index=True, right_index=True)
df_merged

```

```python
# Eliminating rows with more than 2 missing values
df_merged = df_merged[df_merged.isnull().sum(axis=1) < 2]
df_merged
```

```python
# Checking the cities with the most usage according to the bill charges
df_xx = df_merged.sort_values(by = 'VALUE', ascending = False)
df_xx
df_aa = df_xx[['VILLES','VALUE']] [(df_xx['VALUE'] > 250)]
df_aa
```

MERGING df2 and df5 on cell id column


```python
df_merged2 = df2.merge(df5, how='outer', on=["CELL_ID"], left_index=True, right_index=True)
df_merged2
```

```python
# Eliminating rows with more than 2 missing values
df_merged2 = df_merged2[df_merged2.isnull().sum(axis=1) < 2]
df_merged2
```

```python
# Checking the cities with the most usage according to the bill charges
df_yy = df_merged2.sort_values(by = 'VALUE', ascending = False)
df_yy
df_bb = df_yy[['VILLES','VALUE']] [(df_yy['VALUE'] > 250)]
df_bb
```

MERGING df3 and df5 on cell id column

```python
# Rename unmatching columns in df3 to correct format
df13 = df3.rename({'SIET_ID':'SITE_ID', 'CELLID':'CELL_ID'}, axis=1)
df13
```

```python
# merging the 2 datasets
df_merged3 = df13.merge(df5, how='outer', on=["CELL_ID"], left_index=True, right_index=True)
df_merged3
```

```python
# Eliminating rows with more than 2 missing values
df_merged3 = df_merged3[df_merged3.isnull().sum(axis=1) < 2]
df_merged3
```

```python
# Checking the cities with the most usage according to the bill charges
df_zz = df_merged3.sort_values(by = 'VALUE', ascending = False)
df_zz
df_cc = df_zz[['VILLES','VALUE']] [(df_zz['VALUE'] > 250)]
df_cc
```

Combining the 3 data frames with cities with the most usage and ordering them according to billing charge descending order

```python
import pandas as pd
from functools import reduce
data_frames = [df_aa, df_bb, df_cc]
df_mergedxyz = reduce(lambda  left,right: pd.merge(left,right,on=['VILLES'],
                                            how='outer'), data_frames)
df_mergedxyz = df_mergedxyz[df_mergedxyz.isnull().sum(axis=1) < 2]
df_mergedxyz
df99 = df_mergedxyz.groupby(['VILLES']).agg('sum')
df99
df98 = df99.sort_values(by = 'VALUE', ascending = False)
df98
```
